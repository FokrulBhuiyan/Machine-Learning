```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import math
```

```python
data_df = pd.read_csv('test_scores.csv')
data_df.head()
```

```python
data_df['math']
```

```python
def predict_using_sklean():
    df = pd.read_csv("test_scores.csv")
    r = LinearRegression()
    r.fit(df[['math']],df.cs)
    return r.coef_, r.intercept_
```

```python
def gradient_descent(x,y):
    W = b = 0
    l_rate = 0.0001
    m = len(x)
    prev_cost = 0
    for i in range(1000):
        y_pred = W * x + b
        cost = -(1/m)*sum([val**2 for val in (y-y_pred)])
        md = -(2/m)*sum(x*(y-y_pred))
        bd = -(2/m)*sum(y-y_pred)
        W = W - l_rate*md
        b = b - l_rate*bd
        print ('W: {} b: {} cost: {} i: {}'.format(W,b,cost, i))
        if math.isclose(cost, prev_cost, rel_tol=1e-20):
            break
        prev_cost = cost
    return W, b
```

```python
x = np.array(data_df.math)
y = np.array(data_df.cs)

W, b = gradient_descent(x,y)
print("Using gradient descent function: Coef {} Intercept {}".format(W, b))

```

```python
W_sklearn, b_sklearn = predict_using_sklean()
print("Using sklearn: Coef {} Intercept {}".format(W_sklearn,b_sklearn))
```
